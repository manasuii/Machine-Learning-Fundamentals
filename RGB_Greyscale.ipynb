Resizing the image using PIllow library

```python
from PIL import Image
```

```python
img=Image.open('/content/dog.jpg')
img_resized=img.resize((200,200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
img_res=mping.imread('/content/dog_image_resized.jpg')
img_res_plot=plt.imshow(img_res)
plt.show()
```

Covert RGB image to greyscale images using OPenCV

```python
import cv2
```

```python
img=cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
greyscale_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
type(greyscale_image)
```

```python
greyscale_image.shape
```
